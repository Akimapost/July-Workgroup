```python
import numpy as np
import matplotlib.pyplot as plt
import random

n_steps = 1000
steps = np.random.choice([-1, 1], size=n_steps)
position = np.cumsum(steps)
plt.plot(position, color='pink')
plt.title('1D Simple Random Walk')
plt.xlabel('Step')
plt.ylabel('Position')
plt.grid(True)
plt.show()

plt.show()
```

```python
n = 100000
position = np.zeros(n)
for i in range(1, n):
    step = random.choice([-1, 1])
    position[i] = position[i - 1] + step

plt.figure(figsize=(10, 6))
plt.plot(position, color='blue', lw=2, alpha=0.7)
plt.scatter(0, position[0], c='green', s=100, label='Start')  # start point
plt.scatter(n - 1, position[-1], c='red', s=100, label='End')  # end point

plt.title(f'1D Random Walk of n={n} steps', fontsize=16)
plt.xlabel('Step')
plt.ylabel('Position')
plt.grid(True)
plt.legend()

plt.show()
```
